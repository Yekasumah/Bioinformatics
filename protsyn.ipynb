```python
ncov_record = SeqIO.read("sequence.fasta", "fasta")
```

```python

ncov_record
```

```python
ncov_dna = ncov_record.seq
```

```python
ncov_dna
```

Length of sequence

```python
len(ncov_dna)
```

Protein synthesis
converting DNA =>mRNA => protein

```python
#transcription
mrn_ncov = ncov_dna.transcribe()
```

```python
mrn_ncov
```

```python
#Translate to Amino acids
protein_ncov = mrn_ncov.translate()
```

```python
protein_ncov
```

```python
len(protein_ncov)
```

```python
#codon = 3
len(ncov_dna)/3
```

```python
#find all amino acids sequence before stopcodon
aa_ncov = protein_ncov.split("*")
```

```python
aa_ncov
```

```python
clean_ncov = [ i for i in aa_ncov]
```

```python
clean_ncov
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'amino_acids':clean_ncov})
```

```python
df['count'] = df['amino_acids'].str.len()
```

```python
df.head()
```

```python
#largest sequence before *
df.nlargest(10,"count")
```
